```python
cnx = mysql.connector.connect(
    host="localhost",
    user="root",
    password="abc1234!",
    database="northwind"
)

# Bir cursor oluşturun
cursor = cnx.cursor()

# 1. Identify the table that contains the foreign key constraint
# Assume the table is named 'territory'
table_with_foreign_key = "Territory"

# 2. Drop the foreign key constraint
drop_foreign_key_query = f"ALTER TABLE {table_with_foreign_key} DROP FOREIGN KEY territory_ibfk_1"
cursor.execute(drop_foreign_key_query)

# 3. Drop the column in the original table
table_name = "Territory"
column_to_drop = "regionId"
drop_column_query = f"ALTER TABLE {table_name} DROP COLUMN {column_to_drop}"
cursor.execute(drop_column_query)

# Commit the changes
cnx.commit()

# Close the connection
cursor.close()
cnx.close()
```
